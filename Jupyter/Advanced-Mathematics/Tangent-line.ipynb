```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x  = np.linspace(-8, 8, 100)
```

```python
y = 3*x**2 + 4*x + 5
```

```python
plt.plot(x, y)
```

```python
yy = 10*x + 2##(x = 1 处的切线）
```

```python
plt.plot(x, y)
plt.plot(x, yy)
plt.grid()
```

```python
y_minus_2 = -8*x - 7#(-2 处的切线）
```

```python
plt.plot(x, y)
plt.plot(x, yy)
plt.plot(x, y_minus_2)
plt.grid()
```

```python
import numpy as np
import matplotlib.pyplot as plt
x  = np.linspace(-8, 8, 100)
y = 3*x**2 + 4*x + 5
yy = 10*x + 2##(x = 1 处的切线）
y_minus_2 = -8*x - 7#(-2 处的切线）
plt.plot(x, y)
plt.plot(x, yy)
plt.plot(x, y_minus_2)
plt.grid()
plt.title("tangent line")
```
